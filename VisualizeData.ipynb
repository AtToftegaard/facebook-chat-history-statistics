```python
import pandas as pd
import locale
import matplotlib
import matplotlib.pyplot as plt
```

```python
locale.setlocale(locale.LC_ALL, 'da_DK')
df = pd.read_csv("messages.csv", sep=",")

def FormatDate(df):
    df['Datetime'] = pd.to_datetime(df['Date'], errors='coerce' ,format='%d. %b. %Y %H.%M')
    df['Date'] = df['Datetime'].dt.date
    df['DayOfWeek'] = df['Datetime'].dt.weekday_name
    
FormatDate(df)
df.head(100)
```

```python
#Plot who sent how many messages
fig, ax = plt.subplots(figsize=(15,7))
bySender = df.groupby(["Sender"])["Message"].count()
bySender.sort_values(ascending=False).plot.barh()
```

```python
#Plot messages over days of the year
fig, ax = plt.subplots(figsize=(15,7))
byDate = df.groupby(["DayOfWeek"])["Message"].count()
byDate.plot.bar()
```

```python
reactionsdf = pd.read_csv("reactions.csv", sep=",")

def FormatDate(df):
    df['Datetime'] = pd.to_datetime(df['Date'], errors='coerce' ,format='%d. %b. %Y %H.%M')
    df['Date'] = df['Datetime'].dt.date
    df['DayOfWeek'] = df['Datetime'].dt.weekday_name
    
FormatDate(reactionsdf)
reactionsdf.head(100)
```

```python
fig, ax = plt.subplots(figsize=(25,7))
byEmoji = reactionsdf.groupby(["Emoji"])["Emoji"].count()
byEmoji.plot.bar()
```

```python
fig, ax = plt.subplots(figsize=(20,7))
bySender = reactionsdf.groupby(["Sender"])["Emoji"].count()
bySender.plot.bar()
```

```python
fig, ax = plt.subplots(figsize=(20,7))
sursmiley = reactionsdf[reactionsdf.Emoji == '游']
sursmiley.groupby(["Sender"])["Emoji"].count().plot.bar()
```
